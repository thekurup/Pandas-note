```python
#merging
df_customer =pd.DataFrame({
    'id':[4,5,3,7],
    'name':['Tom', 'Jenny', 'James', 'Dan']
})

df_info = pd.DataFrame({
    'id': [2, 3, 4, 5],
    'age': [31, 20, 40, 70],
    'sex': ['F', 'M', 'M', 'F']
})
```

```python
pd.merge(df_customer,df_info)
```

```python
df_order = pd.DataFrame({
    'id': [2,6,4,5],
    'name': ['Jennys', 'James', 'Dan', 'leo'],
    'quantity': [2,4,6,10]
})
df_order
```

```python
pd.merge(df_order,df_customer,on='id')
```

```python
#left join
pd.merge(df_order,df_customer,on='id',how='left')
```

```python
#right join
pd.merge(df_order,df_customer,on='id',how='right')
```

```python
#outer join
pd.merge(df_order,df_customer,on='id',how='outer')
```

```python
#indicator to find merge values are from where
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})

df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})

```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='city',indicator=True)
```
